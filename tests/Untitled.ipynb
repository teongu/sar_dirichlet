```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = [.6, .3, .1]
```

```python
res = np.random.dirichlet(alpha, size=100000)
```

```python
np.sum(res>0.95)
```

```python
plt.hist(res[:,0], bins=20)
plt.show()
```

```python
plt.hist(res[:,1], bins=20)
plt.show()
```

```python
plt.hist(res[:,2], bins=20)
plt.show()
```
